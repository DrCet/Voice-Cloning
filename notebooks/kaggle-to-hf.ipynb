```python
from datasets import Dataset
import datasets
import os
import json
```

```python
data = json.load(open('/kaggle/input/audio-logs-vn-1/logs.json', 'r', encoding='utf-8'))
audio_files = [sample['audio_file'] for sample in data]
texts = [sample['text'] for sample in data]
audio_paths = [os.path.join('/kaggle/input/audio-logs-vn-1/audio_files/audio_files', audio_file) for audio_file in audio_files]
print(audio_files[0])
print(texts[0])
print(audio_paths[0])
```

```python
dataset = Dataset.from_dict({'audio': audio_paths,'text': texts}).cast_column('audio', datasets.Audio(sampling_rate=16000))
dataset = dataset.train_test_split(test_size=0.2)
dataset
```

```python
from IPython import display

print(dataset['train']['text'][0])
print(dataset['train']['audio'][0])
display.Audio(dataset['train']['audio'][0]['array'], rate=16000)
```
