# WhisperTiny

```python
from transformers import pipeline
transcriber = pipeline("automatic-speech-recognition", model="openai/whisper-tiny")
```

```python
audio_file = '/kaggle/input/audio-logs-vn-1/audio_files/audio_files/00-00-15-26-03-2025.wav'
output = transcriber(audio_file, generate_kwargs={"language": "vietnamese"})['text']
print(output)
from IPython import display
display.Audio(audio_file)
```
